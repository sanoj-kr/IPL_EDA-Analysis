### 4.10 Method to know matches won by given team after given date

```python
# method to know matches won by given team after given date

def match_win(team,date):
    winner = df['winner']==team
    date = df['date']>date
    filter = df[winner & date]
    return filter[['season','team1','team2','winner']].head(25)

match_win('Mumbai Indians','2017-01-01')
```

### Top 10 batsman who hits most sixes in IPL

```python
#Top 10 batsman who hits most sixes in IPL
sixes= delivery['batsman_runs']==6
delivery[sixes].groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

### Top 10 batsman who hits most fours in IPL

```python
#Top 10 batsman who hits most fours in IPL

fours= delivery['batsman_runs']==4
delivery[fours].groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

### Top 10 batsman who hits most boundries (fours + sixes) in IPL

```python
#Top 10 batsman who hits most boundries (fours + sixes) in IPL

delivery[fours+sixes].groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

## Most run scored by Virat Kohli against teams

```python
vk = delivery['batsman']=='MS Dhoni'
delivery[vk].groupby('bowling_team')[['batsman_runs','ball']].sum().sort_values('batsman_runs',ascending=False).head()
```

```python
#By using function

def run_scored(batsman_name):
    vk = delivery['batsman']== batsman_name
    return delivery[vk].groupby('bowling_team')[['batsman_runs','ball']].sum().sort_values('batsman_runs',ascending=False).head()
    
```

```python
run_scored('MS Dhoni')
```
